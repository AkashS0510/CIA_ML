```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
coords = np.random.randn(1000, 2)
```

```python
df=pd.DataFrame(coords)
df
```

```python
x=df[0]
y=df[1]
```

```python
plt.scatter(x,y)
```

```python
np.random.seed(0)
slope=np.random.rand()
intercept=np.random.rand()

print(slope)
print(intercept)
```

```python
y_pred=slope*x+intercept
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
```

```python
def cost(y,y_pred):
    c= np.sum((y-y_pred)**2)/(len(y))
    return c
```

```python
cost(y,y_pred)
```

```python
#to update weights:

def update_weights(slope,intercept):
    s=slope-(0.01*(cost(y,y_pred)))
    c=intercept-(0.01*(cost(y,y_pred)))
    return s,c
```

```python
##Gradient Descent

costt=cost(y_pred,y)
l=[]
# print(costt)
for i in range(100):
    slope,intercept=update_weights(slope,intercept)
    y_pred=slope*x+intercept
    
    if costt<cost(y_pred,y):
        break
    costt=cost(y_pred,y)
    l.append(costt)
    print("Cost : ",cost(y_pred,y))
    plt.scatter(x,y)
    plt.plot(x,y_pred,c='r')
    plt.show()
```

```python
plt.plot(l)
plt.title('Gradient Descent Curve');
```

```python
print("Cost: ",cost(y_pred,y))
plt.scatter(x,y)
plt.plot(x,y_pred,c='r')
plt.title("Best Fit Line");
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
x=np.array(x)
x=x.reshape(-1,1)
lr.fit(x,y)
Y_PRED=lr.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,Y_PRED,c='r')
plt.title("Best Fit Line");
```

```python
print("Loss(MSE) using Gradient Descent", cost(y,y_pred))
print("Loss(MSE) using SKlearn", cost(y,Y_PRED))
```

# INFERENCES:

The cost in Gradient Descent Minimizes after nearly 50 iterations.
There is a slight error difference between the mse using GD and Sklearn.
There is a slight differences in the best fit line due to the slight error in the cost function.
